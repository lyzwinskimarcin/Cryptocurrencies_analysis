```python
# Imports
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python

def get_api_data():
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'EUR'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '381ea747-16ca-4e3e-8c7d-68f2f8ca939c',
    }
    
    session = Session()
    session.headers.update(headers)
    
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      # print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)


    df = pd.json_normalize(data['data'])
    df['timestamp'] = pd.to_datetime('now')

    return df
```

```python
def pull_data(pulls_num=333):
    df = get_api_data()
    if not os.path.isfile(r'C:\Marcin Łyżwiński\Projekty portfolio\Web Scraping\Data\APIdata'):
        df.to_csv(r'C:\Marcin Łyżwiński\Projekty portfolio\Web Scraping\Data\APIdata', header=True)
    for i in range(pulls_num-1):
        
        new_df = get_api_data()
        df = pd.concat([df, new_df], ignore_index=True)
        df.to_csv(r'C:\Marcin Łyżwiński\Projekty portfolio\Web Scraping\Data\APIdata', mode='w', header=True)
        print("API Runner completed")
        sleep(60)
    return df
```
